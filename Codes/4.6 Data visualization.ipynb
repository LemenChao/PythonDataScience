## 4.6.4 Changing the value range of the axes of a plot

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x=np.linspace(0,10,100)
plt.rcParams['axes.unicode_minus'] =False
plt.plot(x,np.sin(x))
plt.xlim(11,-2)
plt.ylim(2.2,-1.3)
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,21,-1.6,1.6])
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,21,-1.6,1.6])
plt.axis("equal") 
```

## 4.6.5 Adjusting the margins of a plot

```python
plt.plot(x,np.sin(x))
plt.axis([-1,21,-1.6,1.6])
plt.axis("tight")
```

## 4.6.6 Creating multiple plots on the same coordinates

```python
plt.plot(x,np.sin(x),label="sin(x)")
plt.plot(x,np.cos(x),label="cos(x)")
plt.axis("equal") 
plt.legend() 
```

```python
plt.plot(x,np.sin(x),label="sin(x)")
plt.plot(x,np.cos(x),label="cos(x)")
```

## 4.6.7 Adding an Axes to the current fgure or retrieving an existing Axes

## 4.6.9 Creating more complicate plots

```python
from sklearn.datasets import make_blobs 
X,y=make_blobs(n_samples=300,centers=4,random_state=0, cluster_std=1.0)       
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap="rainbow")
```

## 4.6.10 Data visualization with Pandas

```python
import pandas as pd 
women = pd.read_csv('women.csv',index_col =0)  
women.plot(kind="bar")
plt.show() 
```

```python
women.plot(kind="barh") 
plt.show() 
```

```python
women.plot(kind="bar",x="height",y="weight",color="g") 
plt.show()
```

```python
women.plot(kind="kde")
plt.show()
```

```python
women.plot(kind="bar",x="height",y="weight",color="g") 
plt.legend(loc="best")
plt.show()
```

## 4.6.11 Data visualization with Seaborn

```python
import pandas as pd
import seaborn as sns
sns.set(style="ticks")
df_women = pd.read_csv('women.csv', index_col=0,header=0)
sns.lmplot(x="height", y="weight", data=df_women)
```

```python
sns.kdeplot(women.height, shade=True) 
```

```python
sns.distplot(women.height)  
```

```python
sns.pairplot(women)  
```

```python
sns.jointplot(women.height,women.weight,kind="reg")
```
